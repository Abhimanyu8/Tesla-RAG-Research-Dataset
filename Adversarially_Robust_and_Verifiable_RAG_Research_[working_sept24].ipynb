```
# ==============================================================================
# SECTION 1: SETUP AND INSTALLATIONS
# ==============================================================================
# Installs all necessary libraries for the project.
# ==============================================================================
!pip install transformers datasets sentence-transformers faiss-cpu google-api-python-client google-auth-httplib2 google-auth-oauthlib pandas matplotlib seaborn scikit-learn -q

import torch
from transformers import RagTokenizer, RagRetriever, RagTokenForGeneration, DPRContextEncoder, DPRContextEncoderTokenizer
from datasets import Dataset
import faiss
from sentence_transformers import SentenceTransformer, util
import numpy as np
import os
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from google.colab import drive, auth
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from socket import timeout as TimeoutError

# ==============================================================================
# SECTION 2: CONFIGURATION
# ==============================================================================
# All user-configurable parameters for the research project.
# ==============================================================================
DRIVE_FOLDER_PATH = "RAG Research Dataset"
ADVERSARIAL_DOC_TITLES = ["11", "12"]
DPR_CTX_ENCODER_MODEL = "facebook/dpr-ctx_encoder-single-nq-base"
RAG_TOKEN_MODEL = "facebook/rag-token-nq"

# ==============================================================================
# SECTION 3: UTILITY FUNCTIONS (DATA LOADING & ANALYSIS)
# ==============================================================================
# Core logic for data ingestion, analysis, and visualization.
# ==============================================================================

def mount_drive():
    """Mounts the user's Google Drive to the Colab environment."""
    print("Mounting Google Drive...")
    drive.mount('/content/drive', force_remount=True)
    print("Drive mounted successfully.")

def authenticate_and_build_services():
    """Handles user authentication and builds the API service clients."""
    auth.authenticate_user()
    drive_service = build('drive', 'v3')
    docs_service = build('docs', 'v1')
    return drive_service, docs_service

def read_google_doc(docs_service, document_id):
    """Reads text content from a Google Doc with a retry mechanism."""
    max_retries = 3
    for attempt in range(max_retries):
        try:
            document = docs_service.documents().get(documentId=document_id).execute()
            content = document.get('body').get('content')
            text = ''
            if content:
                for value in content:
                    if 'paragraph' in value:
                        elements = value.get('paragraph').get('elements')
                        for elem in elements:
                            text += elem.get('textRun', {}).get('content', '')
            return text.strip()
        except (HttpError, TimeoutError) as err:
            if attempt < max_retries - 1:
                time.sleep(2 ** attempt)
            else:
                return None

def get_folder_id(drive_service, folder_name_path):
    """Finds the Google Drive ID for a folder path within 'My Drive'."""
    current_folder_id = 'root'
    for part in folder_name_path.split('/'):
        if not part: continue
        query = f"name='{part}' and mimeType='application/vnd.google-apps.folder' and '{current_folder_id}' in parents and trashed=false"
        results = drive_service.files().list(q=query, fields="files(id, name)").execute()
        items = results.get('files', [])
        if not items: return None
        current_folder_id = items[0]['id']
    return current_folder_id

def load_documents_from_drive(drive_service, docs_service, folder_path):
    """Loads all Google Docs from a specified Drive folder."""
    documents = []
    folder_id = get_folder_id(drive_service, folder_path)
    if not folder_id:
        print(f"Error: Folder '{folder_path}' not found in Google Drive.")
        return []
    query = f"mimeType='application/vnd.google-apps.document' and '{folder_id}' in parents and trashed=false"
    results = drive_service.files().list(q=query, fields="files(id, name)").execute()
    items = results.get('files', [])
    for item in items:
        time.sleep(1)
        text_content = read_google_doc(docs_service, item['id'])
        if text_content:
            documents.append({"title": os.path.splitext(item['name'])[0], "text": text_content})
    print(f"Successfully loaded content from {len(documents)} Google Docs.")
    return documents

def generate_answer(query, rag_model, rag_tokenizer):
    """Generates an answer using a given RAG model."""
    inputs = rag_tokenizer(query, return_tensors="pt")
    generated = rag_model.generate(input_ids=inputs["input_ids"])
    return rag_tokenizer.decode(generated[0], skip_special_tokens=True).strip()

def calculate_semantic_shift(base_answer, counterfactual_answer):
    """Calculates semantic shift between two answers using cosine similarity."""
    sbert_model = SentenceTransformer('all-MiniLM-L6-v2')
    base_embedding = sbert_model.encode(base_answer, convert_to_tensor=True)
    counterfactual_embedding = sbert_model.encode(counterfactual_answer, convert_to_tensor=True)
    cosine_sim = util.pytorch_cos_sim(base_embedding, counterfactual_embedding)
    return 1 - cosine_sim.item()

def counterfactual_analysis(query, original_docs_list, baseline_answer, ctx_encoder, ctx_tokenizer, rag_tokenizer):
    """
    Performs a robust counterfactual analysis by instantiating a new RAG model
    for each document ablation.
    """
    influence_scores = {}
    for i in range(len(original_docs_list)):
        current_title = original_docs_list[i]['title']
        counterfactual_docs_list = original_docs_list[:i] + original_docs_list[i+1:]
        if not counterfactual_docs_list: continue

        counterfactual_dataset = Dataset.from_list(counterfactual_docs_list)
        with torch.no_grad():
            counterfactual_dataset = counterfactual_dataset.map(
                lambda ex: {'embeddings': ctx_encoder(**ctx_tokenizer(ex["text"], **tokenizer_kwargs))[0][0].numpy()}
            )
        counterfactual_dataset.add_faiss_index("embeddings")
        counterfactual_retriever = RagRetriever.from_pretrained(RAG_TOKEN_MODEL, indexed_dataset=counterfactual_dataset)

        counterfactual_model = RagTokenForGeneration.from_pretrained(RAG_TOKEN_MODEL, retriever=counterfactual_retriever)

        counterfactual_answer = generate_answer(query, counterfactual_model, rag_tokenizer)
        shift = calculate_semantic_shift(baseline_answer, counterfactual_answer)
        influence_scores[current_title] = shift

        print(f"  - Counterfactual Answer (w/o '{current_title}'): {counterfactual_answer}")
        print(f"  - Influence Score: {shift:.4f}\n")

    return influence_scores

def save_and_plot_results(sorted_scores, query, baseline_answer, filename="rag_analysis_results.csv"):
    """
    Saves the analysis results to a CSV and generates a bar chart.
    Includes a check to handle cases where all scores are zero.
    """
    all_scores_zero = all(score == 0.0 for _, score in sorted_scores)

    if all_scores_zero:
        print("\nRESEARCH FINDING: All influence scores are zero.")
        print("This indicates the model answered from its internal parametric knowledge and ignored all retrieved documents.")
        print("No plot will be generated for this result.")
        results_data = [{"Query": query, "Baseline Answer": baseline_answer, "Ablated Document Title": title, "Influence Score": score, "Note": "Parametric Answer"} for title, score in sorted_scores]
    else:
        results_data = [{"Query": query, "Baseline Answer": baseline_answer, "Ablated Document Title": title, "Influence Score": score, "Note": ""} for title, score in sorted_scores]
        # Plotting logic
        titles = [item[0] for item in sorted_scores]
        scores = [item[1] for item in sorted_scores]
        plt.figure(figsize=(10, 8))
        ax = sns.barplot(x=scores, y=titles, palette="viridis", orient='h')
        ax.set_title(f"Document Influence Scores for Query: '{query}'", fontsize=16)
        ax.set_xlabel("Influence Score (1 - Cosine Similarity)", fontsize=12)
        ax.set_ylabel("Ablated Document Title", fontsize=12)
        for i, score in enumerate(scores):
            ax.text(score + 0.01, i, f"{score:.4f}", va='center', fontweight='bold')
        plt.tight_layout()
        plt.show()

    # Saving logic
    df = pd.DataFrame(results_data)
    mode = 'a' if os.path.exists(filename) else 'w'
    header = not os.path.exists(filename)
    df.to_csv(filename, mode=mode, header=header, index=False)
    print(f"\nCheckpoint created. Results appended to {filename}")

def visualize_document_embeddings(dataset, adversarial_titles):
    """Creates a 2D t-SNE plot of document embeddings."""
    print("\nGenerating t-SNE plot of document embeddings...")
    embeddings = np.array(dataset["embeddings"])
    titles = dataset["title"]
    tsne = TSNE(n_components=2, perplexity=max(1, len(titles)-1), random_state=42, init='pca', learning_rate='auto')
    embeddings_2d = tsne.fit_transform(embeddings)
    colors = ['red' if title in adversarial_titles else '#007ACC' for title in titles]
    plt.figure(figsize=(12, 10))
    plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], c=colors, s=60, alpha=0.8)
    for i, title in enumerate(titles):
        plt.annotate(title, (embeddings_2d[i, 0], embeddings_2d[i, 1]), textcoords="offset points", xytext=(0,10), ha='center')
    plt.title("2D t-SNE Visualization of Document Semantic Space", fontsize=16)
    handles = [plt.Line2D([0], [0], marker='o', color='w', label='Adversarial', markersize=10, markerfacecolor='red'),
               plt.Line2D([0], [0], marker='o', color='w', label='Correct', markersize=10, markerfacecolor='#007ACC')]
    plt.legend(handles=handles, title='Document Type')
    plt.show()

def inspect_retriever(query, rag_model, tokenizer, main_dataset, num_docs=5):
    """
    Inspects the retriever's output by manually replicating the score calculation.
    """
    print(f"\n--- Inspecting Retriever for Query: '{query}' ---")
    inputs = tokenizer(query, return_tensors="pt")
    question_hidden_states = rag_model.rag.question_encoder(inputs["input_ids"])[0]
    question_hidden_states_numpy = question_hidden_states.cpu().detach().numpy()

    retrieved_doc_embeds, doc_ids_array, _ = rag_model.retriever.retrieve(
        question_hidden_states_numpy, n_docs=num_docs
    )

    question_hidden_states_tensor = torch.from_numpy(question_hidden_states_numpy)
    retrieved_doc_embeds_tensor = torch.from_numpy(retrieved_doc_embeds)
    scores_tensor = torch.bmm(question_hidden_states_tensor.unsqueeze(1), retrieved_doc_embeds_tensor.transpose(1, 2)).squeeze(1)
    doc_scores = scores_tensor[0].cpu().numpy()
    doc_ids = doc_ids_array[0]

    for i in range(len(doc_ids)):
        doc_id = doc_ids[i]
        score = doc_scores[i]
        retrieved_doc = main_dataset[int(doc_id)]
        title = retrieved_doc['title']
        text = retrieved_doc['text']
        print(f"\n[Rank {i+1}] Title: '{title}' (Retrieval Score: {score:.4f})")
        print(f"Text: '{retrieved_doc['text'][:250]}...'")
    print("\n--- End of Retriever Inspection ---")

# ==============================================================================
# SECTION 4: INITIALIZATION AND SETUP
# ==============================================================================
mount_drive()
drive_service, docs_service = authenticate_and_build_services()
loaded_docs = load_documents_from_drive(drive_service, docs_service, DRIVE_FOLDER_PATH)

if not loaded_docs:
    raise ValueError("Stopping script: No documents were loaded.")

print("\nInitializing models...")
ctx_tokenizer = DPRContextEncoderTokenizer.from_pretrained(DPR_CTX_ENCODER_MODEL)
ctx_encoder = DPRContextEncoder.from_pretrained(DPR_CTX_ENCODER_MODEL)
rag_tokenizer = RagTokenizer.from_pretrained(RAG_TOKEN_MODEL)
tokenizer_kwargs = {"return_tensors": "pt", "truncation": True, "padding": True, "max_length": 512}

print("Building main knowledge base with FAISS index...")
main_dataset = Dataset.from_list(loaded_docs)
with torch.no_grad():
    main_dataset = main_dataset.map(
        lambda ex: {'embeddings': ctx_encoder(**ctx_tokenizer(ex["text"], **tokenizer_kwargs))[0][0].numpy()}
    )
main_dataset.add_faiss_index("embeddings")

main_retriever = RagRetriever.from_pretrained(RAG_TOKEN_MODEL, indexed_dataset=main_dataset)
main_model = RagTokenForGeneration.from_pretrained(RAG_TOKEN_MODEL, retriever=main_retriever)
print("--- Setup Complete. Ready for analysis. ---")

visualize_document_embeddings(main_dataset, ADVERSARIAL_DOC_TITLES)

# ==============================================================================
# SECTION 5: INTERACTIVE RESEARCH LOOP
# ==============================================================================
while True:
    print("\n" + "="*60)
    print("INTERACTIVE ANALYSIS MENU")
    print("Tip: To properly test the RAG system, ask specific questions that")
    print("     can only be answered by the provided documents, not from general knowledge.")
    print("Good Example: 'What was JB Straubel's role in the Roadster battery pack design?'")
    print("Bad Example:  'Who is the CEO of Tesla?'")
    query = input("\nEnter a query to test (or type 'exit' to quit): ")
    print("="*60)

    if query.lower() == 'exit':
        print("Exiting analysis loop.")
        break

    inspect_retriever(query, main_model, rag_tokenizer, main_dataset)

    print(f"\nAnalyzing query: '{query}'\n")

    baseline_answer = generate_answer(query, main_model, rag_tokenizer)
    print(f"Baseline Answer (using all {len(loaded_docs)} docs): {baseline_answer}\n")
    print("--- Starting Counterfactual Analysis ---\n")

    scores = counterfactual_analysis(query, loaded_docs, baseline_answer, ctx_encoder, ctx_tokenizer, rag_tokenizer)

    print("\n--- Analysis Complete ---")
    print("Influence Scores per Document (sorted high to low):")
    sorted_scores = sorted(scores.items(), key=lambda item: item[1], reverse=True)

    for title, score in sorted_scores:
        print(f"  - '{title}': {score:.4f}")

    if sorted_scores:
        most_influential_doc, high_score = sorted_scores[0]
        print(f"\nMost Influential Document: '{most_influential_doc}' (Score: {high_score:.4f})")

        # This new function handles both plotting and saving robustly.
        save_and_plot_results(sorted_scores, query, baseline_answer)
    else:
        print("\nCould not calculate influence scores.")

```
